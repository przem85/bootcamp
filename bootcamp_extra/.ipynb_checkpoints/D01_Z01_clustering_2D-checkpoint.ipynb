````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN

import numpy.random as rnd
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import fcluster


from sklearn import metrics
````

# Przykład. 
Zastosujmy algorytmy:
 * k-średnie, 
 * DBSCAN, 
 * hierarchiczne klastrowanie 

aby uzyskać najlepszy wynik w sensie:

```python
metrics.adjusted_rand_score
```

````python
rnd.seed(123)
X = np.concatenate([rnd.multivariate_normal((0,0),np.array([[0.1,0],[0,10]]),30),
                    rnd.multivariate_normal((6,5),np.array([[10,0],[0,0.1]]),30),
                    rnd.multivariate_normal((8,-2),np.array([[1,0],[0,1]]),30)],axis=0)
y = np.repeat(range(3),30)

import seaborn as sns
colors = sns.color_palette("Set2", 20)
````

````python
plt.scatter(X[:,0],X[:,1],c=np.array(colors)[y])
plt.ylim(-10,10)
plt.show()
````

### Hierarchical clustering

````python
df = pd.DataFrame(X, index=y)
df

   
row_clusters = linkage(pdist(df, metric='euclidean'), method='ward') 



row_dendr = dendrogram(row_clusters, 
                       labels=y)

plt.tight_layout()
plt.ylabel('Euclidean distance')
plt.show()



part = fcluster(row_clusters, 3, 'maxclust')
plt.scatter(X[:, 0], X[:, 1], c = part)
plt.axis("equal")
plt.show()
````

````python
print(metrics.adjusted_rand_score(y, part))
````

### DBSCAN

````python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps = 1.34, min_samples=3)
labels = model.fit_predict(X)
````

````python
plt.scatter(X[:,0],X[:,1],c=labels+1)
plt.scatter(X[labels==-1,0],X[labels==-1,1], c = 'r', marker = 'x')
plt.show()
````

````python
print(metrics.adjusted_rand_score(y, labels))
````

### k-means

````python
k = 3
km = KMeans(n_clusters=k, n_init=3, init='k-means++')
y_km = km.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_km)
plt.axis("equal")
plt.show()
````

````python
print(metrics.adjusted_rand_score(y, y_km))
````
