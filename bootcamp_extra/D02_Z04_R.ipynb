````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
````

# Szybka transformacja Fouriera

Wygenerujmy dwa sygnały oraz zróbmy ich mieszankę.

````python

Fs = 150.0;  # sampling rate
Ts = 1.0/Fs; # sampling interval
t = np.arange(0,1,Ts) # time vector

ff1 = 5;   # frequency of the signal
ff2 = 14;   # frequency of the signal
y = np.sin(2*np.pi*ff1*t)+np.sin(2*np.pi*ff2*t)


fig, ax = plt.subplots(3, 1,figsize=(10,6))
ax[0].plot(t,np.sin(2*np.pi*ff1*t))
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].plot(t,np.sin(2*np.pi*ff2*t),'r') # plotting the spectrum
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[2].plot(t, y,'g') # plotting the spectrum
ax[2].set_xlabel('Time')
ax[2].set_ylabel('Amplitude')
plt.show()
````

# Zad
Wykonaj 
  * FFT
  * IFFT  
Narysuj pod sobą zmieszany sygnał, wynik FFT oraz odzyskane dane.

````python
n = len(y) # length of the signal
k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
frq = frq[range(np.int(n/2))] # one side frequency range

Y = np.fft.fft(y)/n # fft computing and normalization
z = np.fft.ifft(Y)*n # ifft computing and normalization

Y = Y[range(np.int(n/2))]
````

````python
fig, ax = plt.subplots(3, 1,figsize=(10,6))
ax[0].plot(t,y)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].plot(frq,abs(Y),'r') # plotting the spectrum
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')
ax[2].plot(t, z,'g') # plotting the spectrum
ax[2].set_xlabel('Time')
ax[2].set_ylabel('Amplitude')
plt.show()
````

# Zad
Proszę wykonać podobne zadanie:
 * FFT
 * progowanie
 
$$
 f(x) =
  \begin{cases}
    0       & \quad \text{if } \leq 0.2\\
    x  & \quad \text{if } x > 0
  \end{cases}
$$
 
 * IFFT  
Narysuj pod sobą zmieszany sygnał, wynik FFT oraz odzyskane dane.

````python
YY = np.fft.fft(y)/n # fft computing and normalization
YY[abs(YY)<0.2]=0
z = np.fft.ifft(YY)*n # ifft computing and normalization
YY = YY[range(np.int(n/2))]
````

````python
fig, ax = plt.subplots(4, 1,figsize=(10,6))
ax[0].plot(t,y)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].plot(frq,abs(Y),'r') # plotting the spectrum
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')
ax[2].plot(frq,abs(YY),'r') # plotting the spectrum
ax[2].set_xlabel('Freq (Hz)')
ax[2].set_ylabel('|YY(freq)|')
ax[3].plot(t, z,'g') # plotting the spectrum
ax[3].set_xlabel('Time')
ax[3].set_ylabel('Amplitude')
plt.show()
````
