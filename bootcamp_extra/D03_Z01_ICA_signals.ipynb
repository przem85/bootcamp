```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal

from sklearn.decomposition import FastICA
```

# ICA

## Zad.

Wygeneruj dwa sygna≈Çy.

```python
# #############################################################################
# Generate sample data
np.random.seed(0)
n_samples = 2000
time = np.linspace(0, 8, n_samples)

s1 = np.sin(20 * time)  # Signal 1 : sinusoidal signal
s2 = signal.sawtooth(2 * np.pi * time)  # Signal 2 : square signal

S = np.c_[s1, s2]
```

```python
S /= S.std(axis=0)  # Standardize data

plt.figure()
df = pd.DataFrame(S)
df.plot(subplots=True, figsize=(10, 5));
plt.show()
```

# Zad 

dokonaj mieszania sygna≈Ç√≥w.

```python
# Mix data
A = np.array([[1, -0.7], [0.5, 2]])  # Mixing matrix
A
```

# Zad
Narysuj sygna≈Çy po odmieszaniu.

```python
X = np.dot(S, A.T)  # Generate observations

plt.figure()
df = pd.DataFrame(X)
df.plot(subplots=True, figsize=(10, 5));
plt.show()
```

```python
# Compute ICA
ica = FastICA(n_components=2)
S_ = ica.fit_transform(X)  # Reconstruct signals
A_ = ica.mixing_  # Get estimated mixing matrix
A_
```

# Zad.
 * wygeneruj dwa sygna≈Çy 
 * dokonaj mieszania
 * wykonaj algorytm FastICA
 * narysuj:
     
     * sygna≈Çy oryginalne
     * sygna≈Çy po zmieniszani
     * sygna≈Çy po odmieszaniu
 

```python
# #############################################################################
# Plot results

plt.figure()

models = [X, S, S_]
names = ['Observations (mixed signal)',
         'True Sources',
         'ICA recovered signals']
colors = ['red', 'steelblue', 'orange']

for ii, (model, name) in enumerate(zip(models, names), 1):
    plt.subplot(3, 1, ii)
    plt.title(name)
    for sig, color in zip(model.T, colors):
        plt.plot(sig, color=color)

plt.subplots_adjust(0.09, 0.04, 0.94, 0.94, 0.26, 0.46)
plt.show()
```

 * wygeneruj trzech sygna≈Çy z szumem
 * dokonaj mieszania
 * wykonaj algorytm FastICA
 * narysuj:
     
     * sygna≈Çy oryginalne
     * sygna≈Çy po zmieniszani
     * sygna≈Çy po odmieszaniu
 

```python
# #############################################################################
# Generate sample data
np.random.seed(0)
n_samples = 2000
time = np.linspace(0, 8, n_samples)

s1 = np.sin(2 * time)  # Signal 1 : sinusoidal signal
s2 = np.sign(np.sin(3 * time))  # Signal 2 : square signal
s3 = signal.sawtooth(2 * np.pi * time)  # Signal 3: saw tooth signal


S = np.c_[s1, s2, s3]
S += 0.2 * np.random.normal(size=S.shape)  # Add noise
```

```python
S /= S.std(axis=0)  # Standardize data

plt.figure()
df = pd.DataFrame(S)
df.plot(subplots=True, figsize=(10, 5));
plt.show()
```
