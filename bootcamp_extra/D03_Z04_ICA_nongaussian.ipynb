````python
# Generate sample data
rng = np.random.RandomState(42)
S = rng.standard_t(1.5, size=(20000, 2)) # t-distribution
S[:, 0] *= 2.
````

````python
S /= S.std(axis=0)  # Standardize data

plt.figure(figsize=(8,8))
plt.scatter(S[:, 0], S[:, 1])
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.show()
````

````python
# Mix data
A =  np.array([[1, 1], [0, 2]]) # Mixing matrix
A
````

````python
X = np.dot(S, A.T)  # Generate observations

plt.figure(figsize=(8,8))
plt.scatter(X[:, 0], X[:, 1])
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.show()
````
