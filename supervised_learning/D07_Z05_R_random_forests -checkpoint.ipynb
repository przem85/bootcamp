# Random Forests

````python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
````

````python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.show()
````

````python
plt.hist(y_train)
plt.hist(y_test)
plt.show()
````

````python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag_clf = BaggingClassifier(
    DecisionTreeClassifier(max_leaf_nodes=16, random_state=42),
    n_estimators=500, max_samples=1.0, bootstrap=True, n_jobs=-1, random_state=42)
````

````python
bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)
````

````python
from sklearn.ensemble import RandomForestClassifier

rnd_clf = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, random_state=42)
rnd_clf.fit(X_train, y_train)

y_pred_rf = rnd_clf.predict(X_test)
````

````python
np.sum(y_pred == y_pred_rf) / len(y_pred)  # almost identical predictions
````

````python
# from help_plot import plot_decision_regions
from mlxtend.plotting import plot_decision_regions


plot_decision_regions(X, y, rnd_clf)
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()


plot_decision_regions(X, y, bag_clf)
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()
````

# Ciekawostka - Extra-Trees 


* Gdy budujemy kolejne drzewa  w losowym lesie, w kadym wle rozpatrywany jest losowy podzbir elementw do podziau. 

* Moliwe jest stworzenie drzewa jeszcze bardziej losowego, wykorzystujc losowe podziay dla kadej cechy, zamiast szuka najlepszych moliwych podziau (jak w przypadku zwykych drzew decyzyjnych).

* Taki las losowy nazywa si **Extremely Randomized Trees** (lub **Extra-Trees** w skrcie). 

* Moesz utworzy klasyfikator **Extremely Randomized Trees** za pomoc klasy **ExtraTreesClassifier** z Scikit-Learn. 

* Jego API jest identyczne z klas RandomForestClassifier.
