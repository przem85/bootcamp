# Rozkład Poissona
Rozkład prawdopodobieństwa zmiennej losowej $X$ o rozkładzie Poissona z parametrem $\lambda > 0$ wyraża się wzorem:
$$
P(X = k) = e^{-\lambda} \cdot \frac{\lambda^{k}}{k!}.
$$


```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

# Zadanie

  * narysuj dla niej gęstość i dystrybuantę
  * wylosuj próbkę i narysujesz histogram 
  * narysuj kilka gęstości rozkładu Poissona $\lambda=1,4,10$ 
  * policz skośność i kurtozę dla gęstości Poissona $\lambda=1,4,10$


Zdefiniuj zmienną losową o rozkładzie Poissona  $\lambda=2$

```python
lambda1 = 2
poissonDist = stats.poisson(lambda1)
```

Narysuj gęstość zmiennej losowej o rozkładzie Poissona

```python
fig, ax = plt.subplots(1, 1)
ax.set_xlim([-1,20])
ax.set_ylim([0,1])
x = np.arange(0, 10)

ax.plot(x, poissonDist.pmf(x), 'bo', ms=8, label='poisson pmf')

ax.vlines(x, 0, poissonDist.pmf(x), colors='b', lw=5, alpha=0.5)
ax.legend()
plt.show()
```

Wylosuj próbkę i narysować histogram

```python
trials = poissonDist.rvs(10000)
trials
plt.hist(trials, bins=20)
plt.show()
```

Narysuj dystrybuantę zmiennej losowej o rozkładzie Poissona

```python
p = 0.7
fig, ax = plt.subplots(1, 1)
ax.set_xlim([-2,10])
ax.set_ylim([-0.2,1.2])

x = np.arange(-2, 10,0.01)
ax.plot(x, stats.poisson.cdf(x, lambda1), 'bo--', ms=8, label='bernoulli cdf')

plt.show()
```

Policzy momęnty

```python
mean, var, skew, kurt = stats.poisson.stats(lambda1, moments='mvsk')
print(mean)
print(var)
print(skew)
print(kurt)
```

Narysujmy gętość zmiennej losowej dla różnych prametrów

```python
# Arbitrarily select 3 lambda values
lambdas = [1,2,3,4,5,10]

k = np.arange(20)       # generate x-values
markersize = 8
for par in lambdas:
    plt.plot(k, stats.poisson.pmf(k, par), 'o--',
             label='$\lambda={0}$'.format(par))

# Format the plot
plt.legend()
plt.title('Poisson distribution')
plt.xlabel('X')
plt.ylabel('P(X)')
plt.show()
```
