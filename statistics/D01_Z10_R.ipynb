# Zad. 2
Utwórz nową kolumną o nazwie <tt>HasCabin</tt>, która zawiera 1 jeżeli ktoś miał swoją kabinę i 0 jeśli nie.

```python
df['HasCabin'] = np.nan
df.loc[df.Cabin.notnull(), 'HasCabin'] = 1
df.loc[df.Cabin.isnull(), 'HasCabin'] = 0
df = df.drop(['Cabin'], axis=1)
```

```python
df.HasCabin.value_counts()
```

# Zad. 4
Dokonaj konwrsii danych do postaci numerycznej

```python
embarked = pd.get_dummies(df.Embarked, prefix='Embarked')
embarked
sex = pd.get_dummies(df.Sex, prefix='Sex')
sex
df = df.join(embarked)
df = df.join(sex)
```

```python
vars_to_drop = ['Sex','Embarked']
df = df.drop(vars_to_drop, axis=1)
```
