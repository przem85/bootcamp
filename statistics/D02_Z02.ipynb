# Rozkład dwumianowy
Gdy zmienna losowa $X$ ma rozkład dwumianowy z parametrami $p$ i $n$, zapisujemy go jako $X \thicksim B(n,p)$, a gęstość wyrażona jest wzorem:
$$
P(X = k) = {n \choose k} p^k (1-p)^{n-k},
$$
gdzie
$
{n \choose k} = \frac{n!}{k!(n-k)!}.
$

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

Definiujemy zmienną losową

```python
(p, num) = (0.5, 4)
binomDist = stats.binom(num, p)
binomDist.pmf(np.arange(5))
```

Możemy wylosować próbkę i narysować histogram

```python
trials = binomDist.rvs(100)
trials
plt.hist(trials)
plt.show()
```

Możemy policzyć momenty

```python
(p, num) = (0.5, 4)
mean, var, skew, kurt = stats.binom.stats(num,p, moments='mvsk')
print(mean)
print(var)
print(skew)
print(kurt)
```

Narysujmy gętość zmiennej losowej

```python
fig, ax = plt.subplots(1, 1)
# Display the probability mass function (``pmf``):
ax.set_xlim([-1,5])
ax.set_ylim([0,1])
x = np.arange(0, 5)
ax.plot(x, stats.binom.pmf(x, num, p), 'bo', ms=8, label='bernoulli pmf')
ax.vlines(x, 0, stats.binom.pmf(x, num, p), colors='b', lw=5, alpha=0.5)
plt.show()
```

Narysujmy dystrybuantę zmiennej losowej

```python
fig, ax = plt.subplots(1, 1)
# Display the probability mass function (``pmf``):
ax.set_xlim([-1,5])
ax.set_ylim([-0.2,1.2])

x = np.arange(-5, 5,0.01)
ax.plot(x, stats.binom.cdf(x, num, p), 'bo--', ms=8, label='bernoulli cdf')
plt.show()
```

Narysujmy gętość zmiennej losowej dla różnych prametrów

```python
fig, ax = plt.subplots(1, 1)
# Display the probability mass function (``pmf``):
ax.set_xlim([-1,40])
ax.set_ylim([0,0.2])

x = np.arange(0, 20)
ax.plot(x, stats.binom.pmf(x, 20, 0.5), 'bo--', ms=8, label='n=20, p=0.5', color='r')
x = np.arange(0, 20)
ax.plot(x, stats.binom.pmf(x, 20, 0.7), 'bo--', ms=8, label='n=20, p=0.7', color='g')
x = np.arange(0, 40)
ax.plot(x, stats.binom.pmf(x, 40, 0.5), 'bo--', ms=8, label='n=40, p=0.5', color='b')
ax.legend()
plt.show()
```
