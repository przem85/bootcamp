Napiszmy skrypt w Python, który wygeneruje punkty w kształcie sinusa i je wyrysuje.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.r_[0:10:0.1]
freq = 0.5
x = np.sin(2*np.pi*freq*t)
print(x)
```

```python
# Plot the data
plt.plot(t,x)
# Format the plot
plt.xlabel('Time[sec]')
plt.ylabel('Values')
plt.show()
```
